```python
parameters={}
parameters["W1"]=np.random.randn(4,7)*0.01
parameters["W2"]=np.random.randn(1,4)*0.01
parameters["W3"]=np.random.randn(1,3)*0.01
parameters["b1"]=np.zeros((4,1))
parameters["b2"]=np.zeros((1,1))
parameters["b3"]=np.zeros((1,1))
```

```python
def Linear_function(W,b,X):
    A=np.dot(W,X)+b
    return A
```

```python
def sigmoid(z):
    sigmoid=1/(1+np.exp(-z))
    return sigmoid
```

```python
def relu(z):
    relu=np.maximum(0,z)
    return relu
```

```python
def computed_cost(A,y):
    m=y.shape[0]
    cost=-(1/m)*(np.sum(np.multiply(Y,np.log(A))+np.multiply((1-Y),np.log(1-A))))
    cost = np.squeeze(cost) 
    return cost
```

```python
def Linear_backward(W,dz,A):
    m=Y.shape[0]
    dW=1/m*np.dot(dz,A.T)
    db=1/m*np.sum(dz,axis=1,keepdims=True)
    da=np.dot(W.T,dz)
    return da,dW,db
```

```python
def sigmoid_backward(da,A):
    dz=np.multiply(da,np.multiply(A,(1-A)))
    return dz
```

```python
def Update_gradients(W,b,dW,db,learning_rate):
    W=W-dW*learning_rate
    b=b-db*learning_rate
    return W,b
```

```python
L=len(parameters)
A=[]
A.append(X)
Z=[]
dz=[]
AL=np.array([])
grads={}
costs=[]

for iterations in range(51):
    k=0
    p=0
    m=L//2
    for i in range(1,L+1):
        if i %2!=0:
            k=k+1
            Z.append(Linear_function(parameters["W"+str(k)],parameters["b"+str(k)],A[k-1]))
            print(Z[k-1].shape)

        elif i %2==0 and i!=L:
            A.append(relu(Z[(i//2)-1]))
        elif i==L:
            A.append(sigmoid(Z[-1]))

    Y = Y.reshape(A[-1].shape)        
    cost=computed_cost(A[-1],Y)
    costs.append(cost)
    print(f"Iterration {iterations} cost  {costs[iterations]}")
    dAL=-(np.divide(Y, A[-1]) - np.divide(1 - Y, 1 - A[-1]))
    dz.append(sigmoid_backward(dAL,A[-1]))

    for l in range(L,0,-1):
        if l%2!=0:
            m=(l+1)//2
            grads["dA"+str(m)],grads["dw"+str(m)],grads["db"+str(m)]=Linear_backward(parameters["W"+str(m)],dz[p],A[m-1]) 
            parameters["W"+str(m)],parameters["b"+str(m)]=Update_gradients(parameters["W"+str(m)],parameters["b"+str(m)],grads["dw"+str(m)],grads["db"+str(m)],0.01) 

        if l%2==0 and l!=L:
            dz.append(grads["dA"+str(m)])
            p+=1
```
